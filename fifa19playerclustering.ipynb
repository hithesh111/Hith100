```python
import pandas as pd
import numpy as np
import random
```

```python
fifa19=pd.read_csv('/home/hithesh/Downloads/FIFA 19/data.csv')
features=['Name','Position','Crossing','Finishing','HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']
data=fifa19[features]
data.head()
```

```python
data1=data.loc[:29,:]
```

```python
data1.head()
```

```python
X=data1.iloc[:,2:]
K=4
```

```python
#K-means Algorithm for Clustering

c=[X.iloc[0,:],X.iloc[1,:],X.iloc[2,:],X.iloc[3,:]]
X_class = pd.Series(np.zeros(30))
for j in range(50):
    for i in range(30):
        dist=[0,0,0,0]
        dist[0] = np.linalg.norm(X.iloc[i,:]-c[0])
        dist[1] = np.linalg.norm(X.iloc[i,:]-c[1])
        dist[2] = np.linalg.norm(X.iloc[i,:]-c[2])
        dist[3] = np.linalg.norm(X.iloc[i,:]-c[3])
        
        m=min(dist)
        if(m==dist[0]):
            X_class[i]=0
        elif(m==dist[1]):
            X_class[i]=1
        elif(m==dist[2]):
            X_class[i]=2
        else:
            X_class[i]=3
            
    indices0 = [n for n in range(30) if X_class[n]==0]
    indices1 = [n for n in range(30) if X_class[n]==1]
    indices2 = [n for n in range(30) if X_class[n]==2]
    indices3 = [n for n in range(30) if X_class[n]==3]
    
    c[0]=X.iloc[indices0,:].mean()
    c[1]=X.iloc[indices1,:].mean()
    c[2]=X.iloc[indices2,:].mean()
    c[3]=X.iloc[indices3,:].mean()
```

```python
data1['Class']=X_class
features=['Name','Position','Class']
result=data1[features]
result.head(10)
```

```python
result['Class']=result['Class'].apply(lambda x: 'Messi' if x==0 else x)
result['Class']=result['Class'].apply(lambda x: 'Ramos' if x==1 else x)
result['Class']=result['Class'].apply(lambda x: 'De Bruyne' if x==2 else x)
result['Class']=result['Class'].apply(lambda x: 'De Gea' if x==3 else x)
result
```
